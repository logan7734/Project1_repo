```python
import pandas as pd
import numpy as np
from pandas import ExcelFile
```

```python
ball_data = pd.read_excel("NBA+Data.xlsx", sheetname ="NBA+Data")
```

```python
ball_data.head()
```

```python
ball_data["Year"] = ball_data["Year"].fillna(0)
```

```python
ball_data.head()
```

```python
ball_data["Year"] = ball_data["Year"].astype(int)
```

```python
ball_data.head()
```

```python
#import file with players height and weight. will merge this with the ball_data dataframe
height_weight = pd.read_excel("player_heightweight_data_cleaned.xlsx", sheet_name ="player_data")
```

```python
height_weight.head()
```

```python
list(height_weight)
```

```python
#renaming the first column to match the ball_data
height_weight = height_weight.rename(columns ={'name':'Player'})
height_weight.head()
```

```python
bd_merged = pd.merge(ball_data, height_weight, on = "Player")
```

```python
bd_merged.head()
```

```python
#player height formatted as 'date_time' so it needed to be converted to a string
bd_merged["height"] = bd_merged["height"].astype(str)
```

```python
#after converting, split the original height column into 3 columns on the "-"
bd_merged[["Year(Dropping)","Height (FT)", "Height (In)"]] = pd.DataFrame([x.split("-") for x in bd_merged["height"].tolist()])
```

```python
bd_merged.head()
```

```python
#need to drop the 'year' column from the split becasue it has no relevance and came in formatted incorrectly
bd_merged = bd_merged.drop(columns = ["Year(Dropping)"])
```

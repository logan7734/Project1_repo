```python
#drop N/A values in year: nba_data_df
height_weight_df['weight'] = height_weight_df.weight.fillna(0)

#Convert weight into integer value: height_weight_df
height_weight_df['weight'] = height_weight_df.weight.astype(int)

#Preview
height_weight_df.head()
```

```python
#Merge Both DataFrames on: 'name' column

merged_df = pd.merge(nba_data_df, height_weight_df)
merged_df.head()
```

```python

#remove 'unnamed' and position columns

complete_df = merged_df.drop(["position"],axis=1)

#preview DataFrame: complete_df
complete_df.head()
```

```python

#Filter Years to desired years: we want form 2000 to 2017

my_filter = complete_df.Year >= 1990
desired_years_df = complete_df[my_filter]

#Sort values by 'Years' column

desired_years_df = desired_years_df.sort_values(by=['Year'])

#Reset Index

desired_years_df = desired_years_df.reset_index()

#Delete 'index' column

desired_years_df = desired_years_df.drop(['index'], axis=1)

#preview DataFrame: desired_years_df

desired_years_df.head(2000)
```

```python

#Drop N/A values in 'Pos' column 
desired_years_df["Pos"] = complete_df.Pos.dropna()

#groupby position

position_groups = desired_years_df.groupby('Pos')

#get some information on the groups

position_groups.groups
```

```python

#Get'C' group, Centers; and name it 'centers': use .get_group() method
centers = position_groups.get_group('C')
centers.head()

#group centers by 'year'
centers_year = centers.groupby('Year')
centers_year.groups

#Take a sample year (2000)
two_thousand = centers_year.get_group(2000)

#Get the average pts for centers in the year 2000

avg_pts_2000 = two_thousand
avg_pts_2000.head()
```

```python
#Looking at 2017 as a second sample to do some plotting 


#Get'C' group, Centers; and name it 'centers': use .get_group() method
centers = position_groups.get_group('C')
centers.head()

#group centers by 'year'
centers_year = centers.groupby('Year')
centers_year.groups

#Take a sample year (2000)
two_zero_one_seven = centers_year.get_group(2017)

#Get the average pts for centers in the year 2000

avg_pts_2017 = two_zero_one_seven
avg_pts_2017.head()
```
